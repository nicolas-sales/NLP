```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
porter=PorterStemmer()
sentences=nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
# Finding the post tag

for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[word for word in words if word not in set(stopwords.words("english"))]
    poss_tag = nltk.pos_tag(words)
    print(poss_tag)
```

```python
nltk.pos_tag("Now is the time to make justice a reality for all of God's children".split())
```

```python
# pip install tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

textes = [
    "Maintenant est le bon moment",
    "Faisons de la justice une réalité pour tous les enfants de Dieu",
    "Le temps d'agir, c'est maintenant"
]

# 1) Apprendre le vocabulaire (avec jeton OOV)
tokenizer = Tokenizer(num_words=10000, oov_token="<UNK>")
tokenizer.fit_on_texts(textes)

# 2) Convertir en séquences d'indices
seqs = tokenizer.texts_to_sequences(textes)
print("Séquences :", seqs)

# 3) Padding/Troncature à longueur fixe
max_len = 10
padded = pad_sequences(seqs, maxlen=max_len, padding="post", truncating="post")
print("Padded :\n", padded)

# Vocabulaire (mot -> id)
print("word_index (extrait) :", dict(list(tokenizer.word_index.items())[:10]))
```
