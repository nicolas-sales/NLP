```python
df = df[["v1","v2"]]

df = df.rename(columns={"v1": "label", "v2": "message"})
```

```python
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer # convertit le mot en sa forme canonique
import re
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-z]',' ',df["message"][i]) # remplace tout caractère non alphabétique par un espace
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review]
    review=' '.join(review)
    corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus, df['message']) if i<1]
```

Les caractères qui ne figurent pas dans '[^a-zA-z]'

Donc perte de 3 phrase du corpus pour la suite

map(len, corpus) applique len() à chaque élément de corpus

zip(...) assemble, pour chaque doc, un triplet :

i = longueur du doc (en caractères),

j = doc prétraité (de corpus),

k = message brut (de df['message']).

if i < 1 garde seulement les entrées où la longueur est 0 (docs vides).
Donc on obtient tous les documents vides sous la forme [0, doc_net, doc_brut].

## Liste of sentence, where each sentence is a list of tokens

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

## Vectors average for each sentence

```python
import numpy as np
```

```python
def avg_word2vec(doc):
    vecs = [model.wv.get_vector(w) for w in doc if w in model.wv.index_to_key]
    return np.mean(vecs, axis=0) if vecs else np.zeros(model.wv.vector_size, dtype=np.float32)
    #return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key], axis=0)
```

```python
from tqdm import tqdm
```

```python
# Apply fot the entire sentences
X=[]
for i in tqdm(range(len(words))):
    print(i)
    X.append(avg_word2vec(words[i]))
```
